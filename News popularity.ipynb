```python
# Now we separate the columns into X and Y features labels to build and train our model
# The X features will require 3 inputs the headline, the category , and the day the article was released
headline_feature = ['rate_positive_words','rate_negative_words', 'title_subjectivity', 'title_sentiment_polarity']
category_feature = ['data_channel_is_lifestyle', 'data_channel_is_entertainment','data_channel_is_bus', 'data_channel_is_socmed', 'data_channel_is_tech','data_channel_is_world']
day_feature = ['weekday_is_monday', 'weekday_is_tuesday', 'weekday_is_wednesday', 'weekday_is_thursday', 'weekday_is_friday', 'weekday_is_saturday', 'weekday_is_sunday']
X = df[headline_feature + category_feature + day_feature]
Y = df['popularity']

```

```python
# Attempting to build a simple neural network that would mimic linear regression without the activation layer
import keras
inputs = keras.Input(shape=(len(X.columns),))
outputs = keras.layers.Dense(1)(inputs)
model = keras.Model(inputs=inputs, outputs=outputs)

""" there are 2 ways to compiling a model, one is through using adam optimizer which has its own learning rate and everything already set
while the other way is using RMSprop that gives you options for custom learning rate and other parameters
"""
model.compile(optimizer=keras.optimizers.RMSprop(learning_rate=0.01), loss='mean_squared_error'
              ,metrics= [keras.metrics.RootMeanSquaredError()])

model.summary()
""" Understanding the summary: The input layer has 17 features to train the model, None means that the batch size can vary.
the Dense layer means the output layer. 1 means that there will only be 1 output. Total trainable outputs include the 17 features and 1 bias
because a linear regression equation usually looks like this: y =  m_1 x_1 + m_2 x_2 + ..... + b where we are given the x_n and y. the model
trains to predict the weights (m_n) and the biases (b).
"""
```

```python
# the model will be trained now:

history = model.fit(x=X, y=Y, epochs=50, batch_size= 30)
```

```python
#testing the model.
"""Now to text we need to extract the headline features from raw text using:
1. TextBlob: helps in figuring out subjectivity and sentiment polarity of the headline
2. NLTK: This python api/library will help us in tokenization using the word_tokenize function that
uses the punkt tokenizer to divide a string into substrings. This will also use Opinion Lexicon to give
values for parameters like positive and negative words.
"""
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import opinion_lexicon
nltk.download('punkt') # the tokenizer model
nltk.download('opinion_lexicon') #positive/negative word lists
nltk.download('punkt_tab') # Download the punkt_tab resource
from textblob import TextBlob

def extract_input(headline,category,day):
  tokens = word_tokenize(headline.lower())
  pos_word = 0
  neg_word = 0
  for word in tokens:
    if word in opinion_lexicon.positive():
      pos_word += 1
    elif word in opinion_lexicon.negative():
      neg_word += 1
  rate_positive_words = pos_word / len(tokens)
  rate_negative_words = neg_word / len(tokens)
  title_subjectivity = TextBlob(headline).sentiment.subjectivity
  title_sentiment_polarity = TextBlob(headline).sentiment.polarity
  cat_values = [0,0,0,0,0,0]
  cat_values[category-1] = 1 # Adjusting index for 0-based list
  day_values = [0,0,0,0,0,0,0]
  day_values[day-1] = 1 # Adjusting index for 0-based list
  return [rate_positive_words, rate_negative_words, title_subjectivity, title_sentiment_polarity] + cat_values + day_values
```

```python
print("This project helps to predict how poupualr a news article will be")
headline = input("What is your headline: ")
category = int(input("What categotry of news does it belong to?\n1.lifestyle\n2.entertainment \n3.business \n4.social media\n5.tech \n6.world\nEnter a Number: "))
day = int(input("What day of the week does it fall on?\n1.Monday\n2.Tuesday\n3.Wednesday\n4.Thursday\n5.Friday\n6.Saturday\n7.Sunday\nEnter a Number: "))
user  = extract_input(headline,category,day)
import numpy as np
user_input_array = np.array(user).reshape(1, -1) # Reshape for single prediction
print(model.predict(user_input_array))
```
